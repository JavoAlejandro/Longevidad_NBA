```python
import codecs
# Create the directory if it doesn't exist
if not os.path.exists("request"):
    os.makedirs("request")
    
url = "https://www.basketball-reference.com/draft/NBA_{}.html"

years = list(range(1980, 2023))

count = 0
for year in years:
    if count == 19:
        time.sleep(60)
    response = requests.get(url.format(year))
    with codecs.open(f"request/{year}.html", "w+", encoding="utf-8") as f:
        f.write(response.text)
    count += 1
```

```python
all_stats = []

for year in years:
    with open(f"request/{year}.html", encoding="utf-8") as f:
        page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_='over_header').decompose()
    data = soup.find(id='stats')
    df = pd.read_html(str(data))[0]
    df['year'] = year
    all_stats.append(df)
```

```python
df = pd.concat(all_stats)

df
```
